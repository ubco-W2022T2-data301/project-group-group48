```python
singsong_cleaned = singsong.drop(['audio_mode','liveness', 'time_signature'], axis=1)
singsong_cleaned
#Cleaning data. I'm cleaning by removing these 3 rows that either had 
#no significant variation or were repetetive to other rows
```

```python
singsong_cleaned['loudness_energy'] = singsong_cleaned['loudness'] + singsong_cleaned['energy']
```

```python
singsong_cleaned # this version of my repository has an additional column combining loudness and energy, the two most correlated elements.
```

```python
least_popular = singsong_cleaned.sort_values("song_popularity", ascending=True)
least_popular.head(20) #these are the bottom 20 songs according to popularity
```

```python
most_popular = singsong_cleaned.sort_values("song_popularity", ascending=False)
most_popular.head(10) #these are the top twenty songs according to popularity
```

```python
# Define the combination of elements to group by
group_cols = ["song_name","acousticness", "danceability", "energy", "loudness"]

# Group the songs by the combination of elements and calculate the average popularity
grouped = most_popular.groupby(group_cols)["song_popularity"].mean().reset_index()

# Sort the groups by popularity in descending order
sorted_groups = grouped.sort_values(by="song_popularity", ascending=False)

# Display the top 10 most popular groups
sorted_groups.head(10)
```
