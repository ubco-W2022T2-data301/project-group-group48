```python
import matplotlib.pylab as plt
import seaborn as sns
# Now that I have all genres in place, I will need to make a separate table to get the frequencyof each of the top 23 popular songs and plot that in a graph to visualize the genres that belongto the most popular songs in this dataset. I also combined R&B with R&B/ soul.

data = {'Electronic': 3, 'Latin Urbano': 5, 'Rock': 4, 'Hip-Hop/Rap': 2, 'Alternative/Indie': 2, 'R&B':
2, 'Pop': 2, 'Electropop': 1, 'Rap metal': 1, 'Funk': 1}
df = pd.DataFrame.from_dict(data, orient='index', columns=['frequency'])
df.reset_index(inplace=True)
df.rename(columns={'index': 'Genre'}, inplace=True)
df
```

```python
# Now to plot the bar graph
ax = df.plot(kind='bar', x='Genre', y='frequency', rot=0)
plt.xticks(rotation=45, ha='right')
# This code sets the axis labels and title
ax.set_xlabel('Genre')
ax.set_ylabel('Frequency')
ax.set_title('Genre Frequency Graph')
plt.show()
```
