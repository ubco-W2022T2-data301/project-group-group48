## Processing Data

```python
songdata_sorted = songdata.sort_values(by='song_popularity', ascending=False).reset_index()
songdata_sorted
```

```python
def multiply_columns(row):
    return row['energy'] * row['loudness']

songdata_sorted['Loud_Enrgy_Prod'] = songdata_sorted.aggregate(multiply_columns, axis=1) # This creates a new column that multiplies each loudness value to the corresponding energy value
songdata_sorted
```

### In this processing step, I have ranked my data by song popularity in descending order and created a new variable Loud_Enrgy_Prod that is an aggregate of energy and loudness. I chose to agreegate these two variables because I found in my initaial analysis that loudness and energy  had the biggest correlation 

# Wrangling 

```python
def classify(songdata_sorted):
    songdata_top = (songdata_sorted.head(20)
               .assign(genre=['Reggae', 'Hip Hop', 'Pop', 'Pop', 'Hip Hop', 'Pop', 'Hip Hop', 'Reggae', 'Reggae', 'Pop', 'R&B', 'Pop', 'Pop', 'Pop', 'Rock', 'Pop', 'Rock', 'R&B', 'Rock', 'Rock'])
               .reindex(columns=['song_name', 'genre', 'song_popularity', 'song_duration_ms', 'acousticness', 'danceability', 'energy','instrumentalness', 'key', 'liveness','loudness', 'speechiness', 'tempo', 'audio_valence', 'Loud_Enrgy_Prod']))
    return songdata_top
songdata_top = classify(songdata_sorted)
songdata_top
```

```python
def classify(songdata_sorted):
    songdata_bottom = songdata_sorted.tail(20).sort_values(by=['song_popularity'], ascending = True)\
        .assign(genre=['Latin Pop', 'Son cubano', 'Bolero', 'Hip Hop', 'Rock', 'R&B', 'Rock', 'Rock', 'Rock', 'Rock', 'Rock', 'Rock', 'Rock', 'Rock', 'Rock', 'Salsa', 'Rock', 'Pop', 'R&B', 'Rock'])\
        .reindex(columns=['song_name', 'genre', 'song_popularity', 'song_duration_ms', 'acousticness', 'danceability', 'energy','instrumentalness','key', 'liveness','loudness', 'speechiness', 'tempo', 'audio_valence', 'Loud_Enrgy_Prod'])
    return songdata_bottom

songdata_bottom = classify(songdata_sorted)
songdata_bottom
```

### In this wrangling step, I have created method chains in which I seperate the top 15 popular songs and the least 15 popular songs, reindex them, and I add a new colum that classifies the songs by music genre.

### The visualizations above are scatter plots with best fit lines to show the relationship between my new variable (Loud_Enrgy_Prod) and song popularity. ## Research Question #2

```python
concatenated_df = pd.concat([songdata_top, songdata_bottom], ignore_index=True)
concatenated_df
```

```python
grouped = concatenated_df.groupby('genre')['song_popularity'].mean()
plt.barh(grouped.index, grouped.values)

plt.xlabel('Mean Song Popularity')
plt.ylabel('Genre')
plt.title('Average Song Popularity by Genre')

plt.show()
```

### This bar plot shows the mean song popularity for each genre in the top 15 and bottom 15 songs of the dataset. It is important to note that the samples for some of the genres are small, which means that the results may not be entirely accurate. Additionally, the distribution of genres in the dataset is uneven, with 16 rock songs and 9 pop songs, while only 4 hip hop songs, 4 R&Bs, 3 reggae, and only 1 bolero, 1 latin pop,1 Salsa and 1 son cubano.
### Despite these limitations, the bar plot does show some interesting trends. The highest mean song popularity for both the top 15 and bottom 15 songs was found in the Reggae genre.This is particularly interesting given that there were only 3 samples of this genre.  The Pop genre was a close second followed by the Hip Hop gengre, R&B and Rock respectively. Naturally, since their sample size were much smaller, the Bolero, Latin Pop and Son Cuban genres had the least mean song popularity. 
### Overall, while the sample sizes for some of the genres are small, the bar plot still provides some insights into the relative popularity of different genres in the dataset. However, it is important to interpret these results with caution given the uneven distribution of genres in the dataset. Further analysis with a larger sample size may be necessary to draw more accurate conclusions.

```python
reggae_data = concatenated_df.loc[concatenated_df['genre'] == 'Reggae', ['instrumentalness', 'key','loudness','Loud_Enrgy_Prod']]
pop_data = concatenated_df.loc[concatenated_df['genre'] == 'Pop', ['instrumentalness', 'key','loudness','Loud_Enrgy_Prod']]
hip_hop_data =concatenated_df.loc[concatenated_df['genre'] == 'Hip Hop', ['instrumentalness', 'key','loudness','Loud_Enrgy_Prod']]

# Create a list of the dataframes and the genre labels
data = [reggae_data, pop_data, hip_hop_data]
labels = ['Reggae', 'Pop', 'Hip Hop']

# Create a figure with subplots for each musical element
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

# Loop through each element and plot a box plot for each genre
for i, element in enumerate(['instrumentalness', 'key','loudness','Loud_Enrgy_Prod']):
    row = i // 2
    col = i % 2
    data_to_plot = [genre_data[element] for genre_data in data]
    axs[row, col].boxplot(data_to_plot)
    axs[row, col].set_title(element)
    axs[row, col].set_xticklabels(labels)

plt.tight_layout()
plt.show()
```

### In the scatter plots above, I observe the relationship between the most popular genres (Reggae, Pop and Hip Hop ) and the musical elements that were most correlated or associated with song popularity. As we can see from the plots, all three genres seemed to have somewhat similar average value for the instrumentalness and loud_enrgy_prod. This might suggest that loud_enrgy_prod contributes in a similar manner to the popularity of a musical genre. For loudness, Reggae seemed to have a higher median value than Pop and Hip Hop but most of the samples are below that media. On the other hand, Reggae has the lowest median value for Key and Pop has a couple of extreme outliers even though it is somewhat similar to Hip Hop.
