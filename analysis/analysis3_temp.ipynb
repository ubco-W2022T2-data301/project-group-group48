# Variable Definitions
The are a few variable that I was not too familiar with so I researched their definitions to better understand the role of each in the scope of my analysis:

Acousticness:  a measure used to describe the degree to which a sound or music recording is acoustic or electric in nature. It is a value between 0.0 and 1.0, where 0.0 represents an entirely electric or synthetic sound, and 1.0 represents an entirely acoustic sound.

Speechiness: a measure used to describe the degree to which a sound  contains spoken words, as opposed to singing or instrumental music. It is a value between 0.0 and 1.0, where 0.0 represents music with no speech-like features, and 1.0 represents spoken words with no music.

Audio Valence: Audio valence is a measure used to describe the emotional/ affective content of a sound or music recording. It is a value between 0.0 and 1.0, where 0.0 represents negative or sad emotions, and 1.0 represents positive or happy emotions.

Time Signature: Time signature is a musical notation that specifies the number of beats per measure and the type of note that represents one beat. 

Tempo: Tempo is a measure used to describe the speed or pace of a sound or music recording. Faster tempos generally indicate a more energetic or upbeat musical style, while slower tempos may indicate a more relaxed or contemplative style.
